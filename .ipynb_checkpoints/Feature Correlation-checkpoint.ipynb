```python
%matplotlib inline

import Quandl
import common
import numpy as np
import pandas as pd
import matplotlib.mlab as mlab
import matplotlib.pyplot as plt

from datetime import datetime
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
DJIA = pd.read_csv('../data/DowJonesIA.csv')

opinions = common.load_yahoo_opinions(DJIA)
opinion_sentiments = common.determine_opinion_sentiments(opinions)
prices = common.load_quandl_prices(DJIA, start=datetime(2000, 1, 1))
prices['Close'] = scaler.fit_transform(prices['Close'])
```

# Opinion Correlations

```python
dataset = opinion_sentiments.join(prices['Close']).dropna()

plt.xcorr(dataset['Positive'], dataset['Close'], color='g', usevlines=False, linestyle='-')
plt.xcorr(dataset['Negative'], dataset['Close'], color='r', usevlines=False, linestyle='-')
```

# S&P/ASX 200 Correlations

```python
axjo = Quandl.get("YAHOO/INDEX_AXJO", trim_start=datetime(2000, 1, 1), transformation='rdiff')[['Adjusted Close']]
axjo['AxjoClose'] = scaler.fit_transform(axjo['Adjusted Close'])
axjo.drop('Adjusted Close', inplace=True, axis=1)
# axjo.index = axjo.index - pd.offsets.Day(1)
```

```python
dataset = prices.join(axjo, on=prices.index.get_level_values('Date')).reset_index(level='Ticker', drop=True).dropna()

plt.xcorr(dataset['Close'], dataset['Close'], color='g', usevlines=False, linestyle='-')
plt.xcorr(dataset['AxjoClose'], dataset['Close'], color='r', usevlines=False, linestyle='-')
```

```python
fit = np.polyfit(dataset['Close'],dataset['AxjoClose'],1)
fit_fn = np.poly1d(fit) 

plt.scatter(dataset['Close'], dataset['AxjoClose'], alpha = 0.1, marker='.')
plt.plot(dataset['Close'], fit_fn(dataset['Close']))
```

# SPY Correlation

```python
spy = Quandl.get("GOOG/NYSE_SPY", trim_start=datetime(2000, 1, 1), transformation='rdiff', collapse='monthly')[['Close']]
spy['SpyClose'] = scaler.fit_transform(spy['Close'])
spy.drop('Close', inplace=True, axis=1)
```

```python
dataset = prices.join(spy, on=prices.index.get_level_values('Date')).reset_index(level='Ticker', drop=True).dropna()

plt.xcorr(dataset['Close'], dataset['Close'], color='g', usevlines=False, linestyle='-')
plt.xcorr(dataset['Close'], dataset['SpyClose'], color='r', usevlines=False, linestyle='-')
```
