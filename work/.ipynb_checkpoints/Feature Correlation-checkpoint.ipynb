```python
%matplotlib inline

import Quandl
import common
import pandas as pd
import matplotlib.pyplot as plt

from datetime import datetime
from sklearn.preprocessing import StandardScaler
```

```python
DJIA = pd.read_csv('../data/DowJonesIA.csv')

opinions = common.load_yahoo_opinions(DJIA)
opinion_sentiments = common.determine_opinion_sentiments(opinions)
prices = common.load_quandl_prices(DJIA, start=datetime(2000, 1, 1))
```

```python
scaler = StandardScaler()
```

# Opinion Correlations

```python
dataset = opinion_sentiments.join(prices['Close']).dropna()

plt.xcorr(dataset['Positive'], dataset['Close'], color='g', usevlines=False, linestyle='-')
plt.xcorr(dataset['Negative'], dataset['Close'], color='r', usevlines=False, linestyle='-')
```

# S&P/ASX 200 Correlations

```python
axjo = Quandl.get("YAHOO/INDEX_AXJO", trim_start=datetime(2000, 1, 1), transformation='rdiff')[['Adjusted Close']]
axjo['AxjoClose'] = scaler.fit_transform(axjo['Adjusted Close'])
axjo.drop('Adjusted Close', inplace=True, axis=1)
```

```python
dataset = prices.join(axjo).reset_index(level='Ticker', drop=True)

plt.xcorr(dataset['Close'], dataset['Close'], color='g', usevlines=False, linestyle='-')
plt.xcorr(dataset['Close'], dataset['AxjoClose'], color='r', usevlines=False, linestyle='-')
```
