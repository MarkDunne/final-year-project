# Random vs Real

```python
sample_size = 500

def random_process():
    a = 0
    b = 104
    rho = 0.99
    X, Y = [], []

    aSamples = np.random.normal(size=sample_size)
    bSamples = np.random.normal(size=sample_size)

    for i in range(0, sample_size):
        X.append(i)
        Y.append(a + b)

        a = a * rho + aSamples[i]
        b = b + rho * bSamples[i]

    Y = np.array(Y) - np.mean(Y)
        
    plt.plot(X, Y)

random_process()    
```

# P/B Ratio

```python
pb_returns_abs = pd.read_csv('../data/fundamentals/pb_abs_returns.csv')
```

```python
len(pb_returns_abs['symbol'].unique())
```

```python
dataset = pb_returns_abs[(np.abs(stats.zscore(pb_returns_abs[['pb-abs-ratio']])) < 2).all(axis=1)]
X = dataset['pb-abs-ratio'].reshape(-1, 1)
y = dataset['returns'].reshape(-1, 1)

model = Pipeline([('poly', PolynomialFeatures(degree=1)),
                  ('linear', LinearRegression())])

model.fit(X, y)
xs = np.linspace(-1.75, X.max()+10, 100)


plt.scatter(X, y, color='g', marker='.')
plt.plot(xs, model.predict(xs.reshape(-1,1)), color='k', linewidth=2, label='Best Fit Line')
plt.xlabel('P/B Ratio')
plt.ylabel('Following year growth')

plt.xlim(-1.5, 40)
plt.ylim(-2, 4)
plt.title('Relationship between annual P/B Ratio and following year growth\nComparing 316 companies from 2000 to 2014')
plt.legend()
```
