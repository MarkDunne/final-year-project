# Random vs Real

```python
sample_size = 500

def random_process():
    a = 0
    b = 104
    rho = 0.99
    X, Y = [], []

    aSamples = np.random.normal(size=sample_size)
    bSamples = np.random.normal(size=sample_size)

    for i in range(0, sample_size):
        X.append(i)
        Y.append(a + b)

        a = a * rho + aSamples[i]
        b = b + rho * bSamples[i]

    Y = np.array(Y) - np.mean(Y)
        
    plt.plot(X, Y)

random_process()    
```

```python
start = 0
prices_sample = appl[start : start + sample_size]['Close']
prices_sample = prices_sample - np.mean(prices_sample)

plt.plot(prices_sample)
```

# P/E Ratio Delta

```python
pe_returns_delta = pd.read_csv('../data/fundamentals/pe_delta_returns.csv')
```

```python
len(pe_returns_delta['symbol'].unique())
```

```python
dataset = pe_returns_delta[(np.abs(stats.zscore(pe_returns_delta[['pe-delta-ratio']])) < .5).all(axis=1)]
X = dataset['pe-delta-ratio'].reshape(-1, 1)
y = dataset['returns'].reshape(-1, 1)

model = Pipeline([('poly', PolynomialFeatures(degree=1)),
                  ('linear', LinearRegression())])

model.fit(X, y)
xs = np.linspace(-1.75, 12.75, 100)


plt.scatter(X, y, color='g', marker='.')
plt.plot(xs, model.predict(xs.reshape(-1,1)), color='k', linewidth=2, label='Best Fit Line')
plt.xlabel('P/E Ratio Change')
plt.ylabel('Following year growth')

plt.xlim(-1.5, 8)
plt.ylim(-1.5, 2.5)
plt.title('Relationship between annual relative change in P/E Ratio and following year growth\nComparing 455 companies from 2000 to 2014')
plt.legend()
```

```python
dataset = pe_returns_delta[(np.abs(stats.zscore(pe_returns_delta[['pe-delta-ratio']])) < .5).all(axis=1)]
X = dataset['pe-delta-ratio'].reshape(-1, 1)
y = dataset['returns'].reshape(-1, 1)

high_pe_returns = dataset[stats.zscore(dataset['pe-delta-ratio']) > 1]['returns']
low_pe_returns = dataset[stats.zscore(dataset['pe-delta-ratio']) < -1]['returns']

plt.boxplot([high_pe_returns.values, low_pe_returns.values])
```

# DJIA Window

```python
prices = pd.read_csv('../data/machines/DJIA-window-history.csv')[:10000]
prices = prices.take(np.random.permutation(len(prices)))
```

```python
X = prices.drop(['outcome', 'outcome-class', 'Unnamed: 0'], axis=1)
y = prices['outcome-class']

pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', LogisticRegression())
])

hyperparams = {
    'classifier__penalty': ['l1', 'l2']
}

gs = GridSearchCV(pipe, hyperparams, cv = 10)
scores = cross_val_score(gs, X, y, cv = 10)

scores.mean()
```

```python
pipe = Pipeline([
    ('scaler', StandardScaler()),
    ('classifier', KNeighborsClassifier())
])

hyperparams = {
    'classifier__n_neighbors': np.arange(1, 10),
    'classifier__weights': ['uniform', 'distance']
}

gs = GridSearchCV(pipe, hyperparams, cv = 10)
scores = cross_val_score(gs, X, y, cv = 10)

scores.mean()
```
