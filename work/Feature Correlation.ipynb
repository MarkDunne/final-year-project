# Long Term

```python
def rdiff(series, window):
    f = lambda items: (items[-1] - items[0]) / float(items[0])
    return pd.rolling_apply(series, window, f).dropna()    
```

```python
axjo = Quandl.get("YAHOO/INDEX_AXJO", trim_start=datetime(2000, 1, 1))[['Adjusted Close']]
dji  = Quandl.get("YAHOO/INDEX_DJI" , trim_start=datetime(2000, 1, 1))[['Adjusted Close']]
```

```python
dataset = axjo.join(dji, lsuffix=' axjo', rsuffix=' dji', how='inner')

X = rdiff(dataset['Adjusted Close axjo'], 30)
y = rdiff(dataset['Adjusted Close dji'], 30)
```

```python
plt.acorr(y, color='g', usevlines=False, linestyle='-', maxlags=300)
```

```python
spy  = Quandl.get("GOOG/NYSE_SPY", trim_start=datetime(2000, 1, 1))[['Close']]
prices = rdiff(spy['Close'], 30)
plt.acorr(prices, color='g', usevlines=False, linestyle='-', maxlags=300)
```
