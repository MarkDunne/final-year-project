# Disaster Correlation

http://www.emdat.be/disaster_list/index.html

```python
columns = ['start','end','country','countryISO','location','type','subtype','deaths','affected','damage','disasterid']
disaster_html = open('../data/disaster-table.html').read()
raw_disasters = pd.read_html(disaster_html)[0]
raw_disasters.columns = columns
```

```python
def djia_period_change(disaster):
    start_index = dji.index.searchsorted(disaster.start) - 5
    end_index = dji.index.searchsorted(disaster.end) + 5

    p1 = dji.ix[start_index]
    p2 = dji.ix[end_index]
    
    return (p2 - p1) / p1

disasters = raw_disasters[['start', 'end', 'damage', 'affected', 'deaths']].dropna()
disasters = pd.concat([disasters, disasters.apply(djia_period_change, axis=1)], axis=1)
disasters.rename(columns={'Adjusted Close': 'ValueChange'}, inplace=True)
disasters = (disasters - disasters.mean()) / disasters.std()
```

```python
disasters = disasters[(disasters.damage.abs() < 2) & (disasters.deaths.abs() < 2) & (disasters.affected.abs() < 2)]

plt.bar(1, pearsonr(disasters['damage'], disasters['ValueChange'])[1], label='damage')
plt.bar(2, pearsonr(disasters['deaths'], disasters['ValueChange'])[1], color='g', label='deaths')
plt.bar(3, pearsonr(disasters['affected'], disasters['ValueChange'])[1], color='r', label='affected')
plt.legend()
plt.title('p-value of correlation between disaster\nmetrics and percent change in Dow Jones')
```
