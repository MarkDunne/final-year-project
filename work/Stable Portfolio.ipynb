```python
%matplotlib inline
import numpy as np
import pandas as pd
from datetime import datetime
import pandas.io.data as web
import matplotlib.pyplot as plt
from IPython.display import clear_output
from scipy.interpolate import UnivariateSpline
from scipy.optimize import minimize
```

```python
def random_process(sample_size):
    a = 0
    b = 0
    rho = 1
    X, y = [], []

    aSamples = np.random.normal(size=sample_size)
    bSamples = np.random.normal(size=sample_size)

    for i in range(0, sample_size):
        X.append(i)
        y.append(a + b)

        a = a * rho + aSamples[i]
        b = b + rho * bSamples[i]
    
    spline = UnivariateSpline(X, y)    
    return X, spline(X)
```

```python
X, Ay = random_process(250)
X, By = random_process(250)
X, Cy = random_process(250)

plt.plot(X, Ay, color='r', label='Process A')
plt.plot(X, By, color='g', label='Process B')
plt.plot(X, Cy, color='b', label='Process C')
plt.legend()
```

# Using Scipy minimize

```python
def loss_func(x, y):
    min_dist = np.sqrt((x[0] - x[-1]) ** 2 + (y[0] - y[-1]) ** 2)
    actual_dist = 0.0
    assert len(x) == len(y), "x and y different lengths"
    for i in range(1, len(x)):
         actual_dist += np.sqrt((x[i] - x[i-1]) ** 2 + (y[i] - y[i-1]) ** 2)
    slope = (y[-1] - y[0]) / len(x)
    
    std = np.diff(y).std()
    mov = np.diff(y).mean()
    
    return (actual_dist * std) / (min_dist * slope) + mov   

def loss_func2(weights):
    scaled = (np.array([Ay, By, Cy]).T * weights).T
    y = np.sum(scaled, axis=0)
    error = loss_func(X, y)    
    return error

x0 = np.array([.3, .3, .3])
cons = ({'type': 'eq', 'fun': lambda x: np.sum(x)-1})
res = minimize(loss_func2, x0, method='SLSQP', constraints=cons, options={'disp': True, 'maxiter': 1000})

scaled = (np.array([Ay, By, Cy]).T * res.x).T
best_y = np.sum(scaled, axis=0)

res.x
```

```python
plt.plot(X, Ay, color='r', label='Process A')
plt.plot(X, By, color='g', label='Process B')
plt.plot(X, Cy, color='b', label='Process C')
plt.plot(X, best_y, color='y', label='Optimal')
```

```python
start = datetime(2008, 1, 1)
data = web.DataReader(DJIA.Ticker, 'google', start=start)
```

```python
centered = data.Close - data.Close.ix[0]

for ticker in centered.columns:
    plt.plot(centered[ticker], label=ticker)
```

```python
def loss_func(x, y):
    min_dist = np.sqrt((x[0] - x[-1]) ** 2 + (y[0] - y[-1]) ** 2)
    actual_dist = 0.0
    assert len(x) == len(y), "x and y different lengths"
    for i in range(1, len(x)):
         actual_dist += np.sqrt((x[i] - x[i-1]) ** 2 + (y[i] - y[i-1]) ** 2)
    
    slope = (y[-1] - y[0]) / (x[-i] - x[0])    
    return (min_dist / actual_dist) * slope
```

```python
def find_minima(df, iterations = 100):    
    best_y = None
    best_weights = None
    best_error = 0
    
    for i in range(iterations):
        weights = np.random.beta(.3, .3, size=len(df.columns))
        weights /= weights.sum()
        X = np.arange(len(df))
        y = (df * weights).sum(axis=1)
        error = loss_func(X, y)
        
        if error > best_error:
            best_y = y
            best_error = error
            best_weights = weights    

    return error, best_weights, best_y
            
error, weights, y = find_minima(centered)
```
